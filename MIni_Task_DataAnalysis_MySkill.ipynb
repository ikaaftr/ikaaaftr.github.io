```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load data
df = pd.read_csv('sales_data (1).csv', parse_dates=['Date'])
print("first few rows of the dataset:")
print(df.head)
```

```python
df.dropna()
```

```python
df.drop_duplicates()
```

```python
print("dataset statistics")
print(df.describe)
```

```python
df['Total_Cost'] = df['Cost'] + df['Order_Quantity']
print(df['Total_Cost'])
```

```python
df_canada = df[df['Country']=='Canada']
print(df_canada)
```

```python
#filter row based on a substring match
filtered_df = df[df['Product'].str.contains('Hitch Rack')]
print(filtered_df)
```

```python
selected_columns = df[['Customer_Age', 'Country', "Product_Category"]]

print(selected_columns)
```

```python
#Group data by product category and calculate the total revenue
category_revenue = df_canada.groupby('Product_Category')['Revenue'].sum()
print("\mTotal Revenue by Product Category (Canada)")
print(category_revenue)
```

```python
#Group data by product and calculate the total revenue
Product_revenue = df_canada.groupby('Product')['Revenue'].sum()
print("\nTotal Revenue by Product")
print(Product_revenue)
```

```python
#calculate profit margin
df['Profit_Margin'] = df['Profit'] / df['Revenue']
```

```python
#identify orders with a profit margin greater than 0.2
high_margin_orders = df[df['Profit_Margin'] > 0.2]

print("Orders with Profit Margin greater than 0.2:")
print(high_margin_orders)
```

```python
# Calculate profit margin
df['Profit_Margin'] = df['Profit'] / df['Revenue']

# Identify products with profit margin less than 0.1
low_margin_products = df[df['Profit_Margin'] < 0.1]

# Apply a 10% discount to the Unit_Price of these products
df.loc[df['Product'].isin(low_margin_products['Product']), 'Unit_Price'] *= 0.9

# Display the updated DataFrame
print("Updated DataFrame with Discount Applied:")
print(df)
```

```python
# Calculate sales (Sales) per order
df['Sales'] = df['Unit_Price'] * df['Order_Quantity']

# Calculate total sales per product category
sales_per_category = df.groupby('Product_Category')['Sales'].sum().reset_index()

print("Total Sales per Product Category:")
print(sales_per_category)
```

```python
# Calculate the average revenue for each age group
average_revenue_per_age_group = df.groupby('Age_Group')['Revenue'].mean().reset_index()

print("Average Revenue per Customer Age Group:")
print(average_revenue_per_age_group)
```

```python
# Define the year for which you want to display sales data
year = 2015

# Filter the DataFrame to include only data for the specified year
df_filtered = df[df["Year"] == year]

# Display the sales data for the specified year
print(df_filtered)
```

```python
# Group the sales by country
sales_by_country = df.groupby('Country')['Revenue'].sum()

# Create the pie chart
plt.pie(sales_by_country, labels=sales_by_country.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales by Country')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Display the chart
plt.show()
```

```python
# Create the histogram
plt.hist(df['Customer_Age'], bins=10, edgecolor='black')

# Add labels and title
plt.xlabel('Customer Age')
plt.ylabel('Frequency')
plt.title('Customer Age Distribution')

# Show the plot
plt.show()
```

```python
# Group the data by month and calculate the total revenue for each month
monthly_revenue = df.groupby('Month')['Revenue'].sum()

# Create a bar chart of the monthly revenue
plt.bar(monthly_revenue.index, monthly_revenue.values)
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue')
plt.xticks(rotation=45)
plt.show()
```
